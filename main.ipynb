```python
# This function returns the five cells in the h configuration
# given the parameters.
def update_vertices(a,b,l,h):
    u0.coordinates=[0,0,0]
    u1.coordinates=[a,0,0]
    u2.coordinates=[a/2,np.sqrt(3)*a/2,0]
    v0.coordinates=[0,0,b]
    v1.coordinates=[a,0,b]
    v2.coordinates=[a/2,np.sqrt(3)*a/2,b]
    
    w0.coordinates=[a/2,a/(2*np.sqrt(3)),(b/2)-h]
    w1.coordinates=[a/2,a/(2*np.sqrt(3)),(b/2)+h]

    t0.coordinates=[l*np.cos(np.pi/6),l*np.sin(np.pi/6),b/2]
    t1.coordinates=[a-l*np.cos(np.pi/6),l*np.sin(np.pi/6),b/2]
    t2.coordinates=[a/2,np.sqrt(3)*a/2-l,b/2]

    h0.coordinates=[a/2,a/(2*np.sqrt(3)),-l]
    h1.coordinates=[a/2,a/(2*np.sqrt(3)),b+l]


def update_edges():
    eu01.vertices=[u0,u1]
    eu12.vertices=[u1,u2]
    eu20.vertices=[u2,u0]
    
    ev01.vertices=[v0,v1]
    ev12.vertices=[v1,v2]
    ev20.vertices=[v2,v0]
    
    euv0.vertices=[u0,v0]
    euv1.vertices=[u1,v1]
    euv2.vertices=[u2,v2]

    eu0w.vertices=[u0,w0]
    eu1w.vertices=[u1,w0]
    eu2w.vertices=[u2,w0]
    
    ev0w.vertices=[v0,w1]
    ev1w.vertices=[v1,w1]
    ev2w.vertices=[v2,w1]


    
    ew.vertices=[w0,w1]

    eut0.vertices=[u0,t0]
    eut1.vertices=[u1,t1]
    eut2.vertices=[u2,t2]

    evt0.vertices=[v0,t0]
    evt1.vertices=[v1,t1]
    evt2.vertices=[v2,t2]

    et01.vertices=[t0,t1]
    et12.vertices=[t1,t2]
    et20.vertices=[t2,t0]

    ehu0.vertices=[h0,u0]
    ehu1.vertices=[h0,u1]
    ehu2.vertices=[h0,u2]

    ehv0.vertices=[h1,v0]
    ehv1.vertices=[h1,v1]
    ehv2.vertices=[h1,v2]
    return

def update_polygons():    
    pu01w.edges=[eu01,eu0w,eu1w]
    pu12w.edges=[eu12,eu1w,eu2w]
    pu20w.edges=[eu20,eu2w,eu0w]
    pv01w.edges=[ev01,ev0w,ev1w]
    pv12w.edges=[ev12,ev1w,ev2w]
    pv20w.edges=[ev20,ev2w,ev0w]
    p0w.edges=[euv0,eu0w,ev0w,ew]
    p1w.edges=[euv1,eu1w,ev1w,ew]
    p2w.edges=[euv2,eu2w,ev2w,ew]

    puv01.edges=[eu01,ev01,euv0,euv1]
    puv12.edges=[eu12,ev12,euv1,euv2]
    puv20.edges=[eu20,ev20,euv2,euv0]

    pt.edges=[et01,et12,et20]
    pt0.edges=[eut0,evt0,euv0]
    pt1.edges=[eut1,evt1,euv1]
    pt2.edges=[eut2,evt2,euv2]

    put01.edges=[eu01,et01,eut0,eut1]
    put12.edges=[eu12,et12,eut1,eut2]
    put20.edges=[eu20,et20,eut2,eut0]

    pvt01.edges=[ev01,et01,evt0,evt1]
    pvt12.edges=[ev12,et12,evt1,evt2]
    pvt20.edges=[ev20,et20,evt2,evt0]

    phu01.edges=[ehu0,ehu1,eu01]
    phu12.edges=[ehu1,ehu2,eu12]
    phu20.edges=[ehu2,ehu0,eu20]

    phv01.edges=[ehv0,ehv1,ev01]
    phv12.edges=[ehv1,ehv2,ev12]
    phv20.edges=[ehv2,ehv0,ev20]

def update_polyhedrons():    
    c1.polygons=[phu01,phu12,phu20,pu01w,pu12w,pu20w]
    c2.polygons=[phv01,phv12,phv20,pv01w,pv12w,pv20w]
    d1.polygons=[puv01,pu01w,pv01w,p0w,p1w]
    d2.polygons=[puv12,pu12w,pv12w,p1w,p2w]
    d3.polygons=[puv20,pu20w,pv20w,p2w,p0w]
    shell.polygons=[phu01,phu12,phu20,phv01,phv12,phv20,puv01,puv12,puv20]
    
    #a1.polygons=[put01,put12,put20,pt,pu]
    #a2.polygons=[pvt01,pvt12,pvt20,pt,pv]
    #b1.polygons=[pt0,put01,pvt01,pt1]
    #b2.polygons=[pt1,put12,pvt12,pt2]
    #b3.polygons=[pt2,put20,pvt20,pt0]
    return

def update_all(a,b,l,h):
    update_vertices(a,b,l,h)
    update_edges()
    update_polygons()
    update_polyhedrons()
    return
    
```

```python
def area():
    print(c1.surface_area())
    print(c2.surface_area())
    print(d1.surface_area())
    print(d2.surface_area())
    print(d3.surface_area())
    print(" ")
    #total=c1.surface_area()+c2.surface_area()+d1.surface_area()+d2.surface_area()+d3.surface_area()
    #print("Total Area=",total)
    return
    

    
def volumes():
    a=(round(c1.volume(),3))
    b=(round(c2.volume(),3))
    c=(round(d1.volume(),3))
    d=(round(d2.volume(),3))
    e=(round(d3.volume(),3))
    if(a==b and a==c and a==d and a==e):
        print("Volumes check")
        print("Total Volume=",a+b+c+d+e)

    else:
        print("volume warning!")
        print(a,b,c,d,e)
```

```python
def vol_fit(a,b,l,h,epsilon,target_volume):
    update_all(a,b,l,h)
    
    test_pyramid=d1.volume()-target_volume
    while (test_pyramid)>epsilon:
        b-=0.001
        update_all(a,b,l,h)
        test_pyramid=d1.volume()-target_volume
        #print(test_pyramid)

    #print("pyramid fitting done")
    test_tetrahedron=target_volume-c1.volume()
    while (test_tetrahedron)>epsilon:
        l+=0.001
        update_all(a,b,l,h)
        test_tetrahedron=target_volume-c1.volume()
    #print("tetrahedron fitting done")


    return
```
